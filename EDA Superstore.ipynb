## üß≤ Univariate Analysis 

### 1Ô∏è‚É£ - Categorical Analysis

```python
categorical_cols = ['Customer Segment','Category','Region','Payment Mode','Delivery Status','Auto Reorder']

for col in categorical_cols:
    print(f"*--- {col} ---*")
    print(f"{df[col].value_counts()}")


    #Visualization üìä
    plt.figure(figsize = (4,4))
    df[col].value_counts().plot(kind = 'bar', color = 'skyblue')
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation = 0)
    plt.tight_layout()

    plt.show()
```

### 2Ô∏è‚É£ - Numerical Analysis

```python
numerical_cols = ['Quantity','Unit Price','Discount (%)','Sales Amount','Cost Price','Profit','Stock Left','Reorder Quantity']

for col in numerical_cols:
    print(f"*--- {col} ---*")
    print(f" mean    : {df[col].mean():.2f}")
    print(f" median  : {df[col].median():.2f}")
    print(f" std dev : {df[col].std():.2f}")
    print(f" min     : {df[col].min():.2f}")
    print(f" max     : {df[col].max():.2f}")

    #Visualization üìä
    plt.figure(figsize = (8,4))

    plt.subplot(1,2,1)
    df[col].hist(bins = 30, color = '#FFD700', alpha = 0.7)
    plt.title(f"Distribution of {col}")

    plt.subplot(1,2,2)
    df.boxplot(column = col)
    plt.title(f"Box-Plot of {col}")

    plt.tight_layout()
    
    plt.savefig("numerical_hist_boxplot.pdf")

    plt.show()
    
```

## üÖ±Ô∏è - Bivariate Analysis 

```python
#Category Performance


print("*=== Sales & Profit by Category ===*")

category_performance = df.groupby('Category').agg({'Sales Amount' : 'sum', 'Profit' : 'sum', 'Order ID' : 'count'}).rename(columns = {'Order ID' : 'Order Count'})

category_performance['Profit Margin'] = round((category_performance['Profit']/category_performance['Sales Amount'])*100,2)

print(category_performance,"\n\n")


# Visualization : üìä

fig, axes = plt.subplots(2, 2, figsize = (15,12))

#sales by Category :
category_performance['Sales Amount'].plot(kind = 'bar', alpha = 0.8, color = 'blue', ax= axes[0,0])
axes[0,0].set_title("Total Sales By Category")
axes[0,0].tick_params(axis='x', rotation=30)

#Profit by Category :
category_performance['Profit'].plot(kind = 'bar', ax= axes[0,1], color = 'green', alpha = 0.8)
axes[0,1].set_title("Total Profit By Category")
axes[0,1].tick_params(axis='x',rotation=30)

#Profit Margin by Category :
category_performance['Profit Margin'].plot(kind = 'bar', alpha = 0.8, ax = axes[1,0], color = 'orange')
axes[1,0].set_title("Profit Margin By Category")
axes[1,0].tick_params(axis='x', rotation=30)

#Order Count by Category : 
category_performance['Order Count'].plot(kind = 'bar', alpha = 0.8, color = 'red', ax = axes[1,1])
axes[1,1].set_title("Order Count By Category")
axes[1,1].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```

```python
#Regional Performance

regional_performance = df.groupby('Region').agg({'Sales Amount' : 'sum', 'Profit': 'sum', 'Order ID': 'count'}).rename(columns = {'Order ID': 'Order Count'})

regional_performance['Profit Margin'] = round((regional_performance['Profit']/regional_performance['Sales Amount'])*100,2)

print(regional_performance,'\n\n')

# visualization üìä
plt.figure(figsize=(12,8))

#profit By Region
plt.subplot(2,2,1)
regional_performance['Profit'].plot(kind = 'pie',autopct = '%1.1f%%')
plt.title("Total Profit By region")

#Sales Amount By Region
plt.subplot(2,2,2)
regional_performance['Sales Amount'].plot(kind = 'pie',autopct = '%1.1f%%')
plt.title("Sales Amount by Region")

#Profit Margin By Region
plt.subplot(2,2,3)
sns.barplot(x=regional_performance.index,y='Profit Margin', data = regional_performance,hue = 'Region')
plt.title("Profit Margin By Region")
plt.xticks(rotation = 0)

#Order Count By Region
plt.subplot(2,2,4)
sns.barplot(x=regional_performance.index,y='Order Count', data = regional_performance, hue = 'Region')
plt.title("Order Count By Region")
plt.xticks(rotation = 0)

plt.tight_layout()

plt.savefig("Bivariate_pie_bar.pdf")

plt.show()
```

### ‚åõ Time Series Analysis 

```python
print("*=== Time Series Analysis ===*")

df['Order Month'] = df['Order Date'].dt.to_period('M')
df['Order Year'] = df['Order Date'].dt.year

monthly_trends = df.groupby('Order Month').agg({'Sales Amount' : 'sum','Profit':'sum', 'Order ID': 'count'}).rename(columns = {'Order ID':'Order Count'})

print("Monthly Trends,'\n\n'")
print(monthly_trends.tail(10),'\n\n')

#Visualization üìä
plt.figure(figsize =(15,10))

#Sales trends
plt.subplot(3,1,1)
monthly_trends['Sales Amount'].plot(kind = 'line', color = 'red', marker = '*')
plt.title("Monthly Sales Trend")
plt.ylabel('Sales Amount')
plt.grid(True)

#Profit
plt.subplot(3,1,2)
monthly_trends['Profit'].plot(kind = 'line', color = 'green', marker = "*")
plt.title("Monthly Profit Trend")
plt.ylabel('Profit')
plt.grid(True)

#Order Count
plt.subplot(3,1,3)
monthly_trends['Order Count'].plot(kind = 'line', color = 'lightblue', marker = "*")
plt.title("Monthly Order Count Trend")
plt.ylabel('Order Count')
plt.grid(True)

plt.tight_layout()

plt.savefig("line_chart.pdf")

plt.show()
```

## Advanced Analysis

```python
print("*=== Correlation Analysis ===*")

numerical_df = df[['Profit','Cost Price', 'Sales Amount','Discount (%)', 'Unit Price','Quantity']]

correlation_matrix = numerical_df.corr()
print("*== Correlation Matrix ==*")
print(correlation_matrix,'\n\n')

#Visualization üìä

plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix , annot = True, cmap = 'coolwarm', center = 0,square = True, linewidths= 0.5)
plt.title("Correlation Heatmap")

plt.tight_layout()
plt.savefig("heatmap_chart.pdf")

plt.show()
```

## Customer Segment Analysis

```python
print("*=== Customer Segment Analysis ===*")

segment_analysis = df.groupby('Customer Segment').agg({'Sales Amount' : ('sum','mean'), 'Profit' : ('sum','mean'), 'Order ID' : 'count', 'Discount (%)': 'mean'}).round(2)

segment_analysis.columns = ['Total Sales', 'Avg Sales','Total Profit','Avg Profit', 'Order Count', 'Avg Discount']
print(segment_analysis,'\n\n')

#Visualization üìä

fig, axes =plt.subplots(2,2,figsize=(15,10))

segment_analysis['Total Sales'].plot(kind = 'bar',ax=axes[0,0],color = 'lightblue')
axes[0,0].set_title("Total Sales By Customer Segment")
axes[0,0].tick_params(axis = 'x',rotation = 0)

segment_analysis['Total Profit'].plot(kind = 'bar',ax=axes[0,1],color = 'black')
axes[0,1].set_title("Total Profit By Customer Segment")
axes[0,1].tick_params(axis = 'x',rotation = 0)

segment_analysis['Avg Profit'].plot(kind = 'bar',ax=axes[1,0],color = 'red')
axes[1,0].set_title("Avg Profit By Customer Segment")
axes[1,0].tick_params(axis = 'x',rotation = 0)

segment_analysis['Avg Discount'].plot(kind = 'bar',ax=axes[1,1],color = 'gold')
axes[1,1].set_title("Avg Discount By Customer Segment")
axes[1,1].tick_params(axis = 'x',rotation = 0)

plt.tight_layout()
plt.show()
```

## Key Insight & Summary

```python
print("*=== Key Insights & Summary ===*\n\n")

total_profit = df['Profit'].sum()
total_sales = df['Sales Amount'].sum()
total_orders = df['Order ID'].nunique()
avg_profit_margin = (total_profit/total_sales)*100

print("OverAll Performance: \n")

print(f"Total Sales : ${total_sales:,.2f}")
print(f"Total Profit : ${total_profit:,.2f}")
print(f"Total Orders : {total_orders}")
print(f"Avg Profit Margin: {avg_profit_margin:.2f}%\n\n")

#Top Performaning Category

top_categories = df.groupby('Category')['Profit'].sum().sort_values(ascending = False)
print("Top Performing Categories \n")

for i , (Category,Profit) in enumerate(top_categories.items(),1):
    print(f"{i} . {Category}: ${Profit:,.2f}")

best_region = df.groupby('Region')['Profit'].sum().idxmax()
print(f"\n\nBest Region :\n \n{best_region}: {Profit}\n\n")

delivery_stats=df['Delivery Status'].value_counts(normalize = True)*100
print("Delivery Status\n")
for status, percentage in delivery_stats.items():
    print(f"{status} : {percentage:.1f}%")


best_segment = df.groupby('Customer Segment')['Profit'].sum().idxmax()
print(f"\n\nMost Profitable Customer Segment : {best_segment}")
```

## Export Results

```python
analysis_results= {'total_sales' : total_sales,'total_profit' : total_profit,
                   'total_orders': total_orders,'avg_profit_margin' : avg_profit_margin,
                   'top_categories' : top_categories,'best_region': best_region,'best_segment': best_segment}

#Convert to DataFrame and Save
results_df = pd.DataFrame([analysis_results])
results_df.to_csv('eda_analysis_summary_ssma.csv', index = False)
print("\nAnalysis Summary Successfully Saved ‚úÖ to eda_analysis_summary_ssma.csv ")
```
